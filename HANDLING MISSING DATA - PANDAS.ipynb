```python
import pandas as pd
df=pd.read_csv('weather_data.csv',parse_dates=['day'])
df
```

# 1st Method using fillna function

```python
df.fillna(0)
#Fills the missing value with the passed value as 0.
```

```python
#In this method we pass the values to replaced at a particular column using the Dict.
new_df=df.fillna({
    'temperature':0,
    'windspeed':0,
    'event':'no event '
})
new_df
```

```python
#the ffill method fills the previous value forwardly to the next data missing space
#The limit function limits the filling of the missing value as seen in the Output.
new_df1=df.fillna(method="ffill",limit=1)
new_df1
```

```python
#Similarly to the prev case we can use the bfill that is the backward fill method
new_df1=df.fillna(method="ffill")
new_df1
```

# 2nd Method is the Interpolate Method

```python
#The interpolate method fills the approximation data on its own.
#this method is considered to be the better guess for the missing values.
new_df1=df.interpolate()
new_df1
```

# 3rd Method Dropna

```python
#This function basically drops all the mising values on the dataframe.
new_df=df.dropna()
new_df
```

```python
#the thres method or the parameter holds the data that has atleast one Non missing data
new_df=df.dropna(thresh=1)
new_df
```

# 4th method using replace function

```python
#This method takes2 parameter .The first is the value to be replaced and the second parameter is the value that needs to be replaced.
import numpy as np
new_df=df.replace(28.0,value=np.NaN)
new_df
```

```python
#replacing multiple values
new_df=df.replace([28.0,32.0],np.NaN)
new_df
```

```python
#Replacing per column
new_df=df.replace({
    'windspeed':9.0,
    'event':'sunny'
},np.NaN)
new_df
```

```python
#replacin using mapping 
ew_df = df.replace({
        -99999: np.nan,
        'no event': 'Sunny',
    })
new_df
```
